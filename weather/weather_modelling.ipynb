```python
import pandas as pd
import numpy as np
import pytz
from datetime import datetime
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt

def clean():
    est = pytz.timezone('US/Eastern')
    utc = pytz.utc
    fmt = '%Y-%m-%d %H:%M:%S %Z%z'
    fmt2 = '%Y-%m-%d %H:%M:%S'

    weather = pd.read_csv("weather_history.csv")
    weather["temp"] = weather["temp"] - 273
    weather["temp_min"] =  weather["temp_min"] -273
    weather["temp_max"] =  weather["temp_max"] -273

    weather["dt"] = pd.Series([utc.localize(datetime.utcfromtimestamp(i)).astimezone(est).strftime(fmt2) for i in weather["dt"]])

    clean = weather[['dt','temp','humidity','wind_speed','clouds_all','weather_main','weather_description']]


    clean.to_csv("clean_weather.csv")
```

```python
weather = pd.read_csv("clean_weather.csv")
date_time = weather["dt"].str.split(" ", n=-1, expand=True)
    
month = date_time[0].str.split("-",n=-1, expand=True)[1].astype(int)
hour = date_time[1].str.split(":", n=-1, expand=True)[0].astype(int)

weather["Year"] = date_time[0].str.split("-",n=-1, expand=True)[0].astype(int)
weather["Month"] = month
weather["Day"] = date_time[0].str.split("-",n=-1, expand=True)[2].astype(int)
weather["Hour"] = hour

weather["Morning"] = np.where((hour >= 5) & (hour <=10), 1, 0)
weather["Afternoon"] = np.where((hour >= 11) & (hour <=16), 1, 0)
weather["Evening"] =np.where((hour >= 17) & (hour <=22), 1, 0)
weather["Night"] = np.where((hour <=4) | (hour ==23), 1, 0)

weather["Spring"] = np.where((month >= 3) & (month <=5), 1, 0)
weather["Summer"] = np.where((month >= 6) & (month <=8), 1, 0)
weather["Fall"] = np.where((month >= 9) & (month <=11), 1, 0)
weather["Winter"] = np.where((month == 12) | (month <=2), 1, 0)

onehot_encoder = OneHotEncoder(sparse=False,categories='auto')
data = weather["weather_main"].values.reshape(-1,1)
onehot_encoder.fit(data)
features = onehot_encoder.categories_[0].tolist()

onehot_encoded = pd.DataFrame(onehot_encoder.fit_transform(data),dtype="int",columns=features)

new = pd.concat([weather, onehot_encoded], axis= 1)

Header = ["Year","Month","Day","Hour","temp","humidity","wind_speed","clouds_all","Morning","Afternoon","Evening","Night","Spring","Summer","Fall","Winter"]
Header.extend(features)
new = new[Header]
new.to_csv("feature_weather.csv")
```

# Sampling Imbalanced Data
Due to data imbalance, we sample from **clear** and **rain** days for training.

```python
#sample from 
rained = joined[joined['Rain']==1] #538
clear = joined[joined['Clear']==1] #1636

num_sample = [50, 100, 150, 200, 250, 300, 350, 400, 450, 500]

train_features = ["temp","humidity","wind_speed","clouds_all","Morning","Afternoon","Evening","Night",
                  "Spring","Summer","Fall","Winter"]
train_features.extend(features)

averaged_loss = []

for i in num_sample:
    
    sampled_rain = rained.sample(n=i)
    sampled_clear = clear.sample(n=i)
    
    sampled = sampled_rain.append(sampled_clear)
    
    X = sampled[train_features]
    y = sampled[["total ride count"]]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

    regression_model = LinearRegression()
    regression_model.fit(X_train, y_train)


    y_predict = regression_model.predict(X_test)
    regression_model_mse = mean_squared_error(y_predict, y_test)
    averaged_loss.append(regression_model_mse/i)
    plotting(y_test, y_predict, False)
```

```python
import pandas as pd
import numpy as np
import pytz
from datetime import datetime
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


def test_training(X, y):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

    regression_model = LinearRegression()
    regression_model.fit(X_train, y_train)
    y_predict = regression_model.predict(X_train)
    regression_model_mse = mean_squared_error(y_predict, y_train)

    print (regression_model_mse)
    return regression_model, X_train, X_test, y_train, y_test, y_predict

#'Clear', 'Clouds', 'Fog', 'Haze', 'Mist', 'Rain', 'Snow', 'Thunderstorm'
joined = pd.read_csv('taxi_joined_2015.csv')
train_features = ["temp","humidity","wind_speed","clouds_all","Morning","Afternoon","Evening","Night",
                  "Spring","Summer","Fall","Winter",]
plt.figure(figsize=(15, 8))
color_code = {'Clear':'r',
            'Rain':'g',
             'Snow':'b'}
for i in color_code:
    X = joined[(joined[i]==1)][train_features]
    y = joined[(joined[i]==1)][["Number of rides"]]
    regression_model, X_train, X_test, y_train, y_test, y_predict = test_training(X, y)
    plt.scatter(y_train, y_predict, color=color_code[i], label=i)

plt.plot(y_test, y_test)
plt.xlabel('y_test'); plt.ylabel('y_predict')
#plt.title()
plt.legend()
plt.show()
```
